```python
import pandas as pd
import sqlite3
from IPython.display import display, Markdown, Latex

import warnings
warnings.filterwarnings('ignore')
```

```python
"""
The tables include
Answer
Question
Survey
"""
con = sqlite3.connect("./mental_health.sqlite")
cur = con.cursor()
```

```python
question_df = pd.read_sql_query("SELECT * from Question", con)
answer_df = pd.read_sql_query("SELECT * from Answer", con)
```

```python
questions = "**Survey Questions:** \n"
for i in range(len(question_df)):
     questions += str(question_df.questionid[i]) + ". " + question_df.questiontext[i] + "\n"
display(Markdown(questions))
```

```python
query = """
        SELECT *
        FROM Answer 
        left outer join
        (
            SELECT * from Question
        )questions
        on Answer.QuestionID=questions.questionid
        """
question_and_answer_df = pd.read_sql_query(query, con)
```

```python
responses = []
for user in question_and_answer_df.UserID.unique():
    user_df = question_and_answer_df[question_and_answer_df['UserID']==user]
    obj = {}
    obj['user id']=user
    obj['survey year']=max(user_df["SurveyID"])
    for question in user_df.questiontext:
        obj[question]=user_df['AnswerText'][user_df['questiontext']==question].values[0]
    responses.append(obj)
```

```python
responses_df = pd.DataFrame(responses)
```

```python
responses_df["What is your gender?"] = responses_df["What is your gender?"].apply(lambda x: str(x).lower())
```

```python
country_df = responses_df[["What country do you live in?",'user id']].\
groupby(["What country do you live in?"]).count().reset_index()
```

```python
country_df.columns = ['What country do you live in?','responses']
```

```python
responses_df['Mental Health'] = ''
responses_df['Mental Health'][responses_df["Do you currently have a mental health disorder?"]=='Yes']='Yes'
responses_df['Mental Health'][responses_df["Have you had a mental health disorder in the past?"]=='Yes']='Yes'
responses_df['Mental Health'][responses_df['Mental Health']!='Yes']='No'

ever_mental_health_df= responses_df[["Mental Health",
                           'user id']].\
groupby(["Mental Health"]).count().reset_index()
ever_mental_health_df
```

```python
fig = px.bar(ever_mental_health_df, x="Mental Health", 
             y="user id")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Pie(labels=ever_mental_health_df["Mental Health"], 
                             values=ever_mental_health_df["user id"], hole=.65)])
fig.show()
```

```python
mental_country_df = responses_df[responses_df['Mental Health']=='Yes'
                                ][["What country do you live in?",'user id']].\
groupby(["What country do you live in?"]).count().reset_index()
mental_country_df.columns = ['What country do you live in?','responses that had mental health']
```

```python
mental_country_df = mental_country_df.merge(country_df, on='What country do you live in?')
```

```python
mental_country_df['ratio of responses with mental health problems'] = \
mental_country_df['responses that had mental health']*100/mental_country_df['responses']
```

```python
mental_country_df=mental_country_df[mental_country_df['responses']>=10].sort_values(
    by='ratio of responses with mental health problems', ascending=False)
```

```python

fig = px.bar(mental_country_df, x='What country do you live in?', 
             y='ratio of responses with mental health problems',
             color='ratio of responses with mental health problems'
             , height=600)
fig.show()
```

```python
gender_df = responses_df[['Mental Health','What is your gender?','user id']].groupby(['What is your gender?','Mental Health']).count().\
reset_index().sort_values(by='user id',ascending=False)
```

```python
position_health_df= responses_df[["Which of the following best describes your work position?",
                           'user id','Mental Health']].\
groupby(["Which of the following best describes your work position?",'Mental Health']).count().\
reset_index().sort_values(by='user id',ascending=False)
position_health_df
fig = px.bar(position_health_df, x="Which of the following best describes your work position?", 
             y="user id", color="Mental Health",
             pattern_shape="Mental Health", pattern_shape_sequence=["x", "|"], height=600)
fig.show()
```

```python
position_health_ratio_df=position_health_df.pivot(index='Which of the following best describes your work position?', 
                         columns='Mental Health', values='user id').reset_index()
position_health_ratio_df['yes ratio'] = ''
position_health_ratio_df['yes ratio'] = position_health_ratio_df['Yes']*100/(position_health_ratio_df['Yes']+position_health_ratio_df['No'])

position_health_ratio_df = position_health_ratio_df.sort_values(by='yes ratio',ascending=False)
```

```python
fig = px.bar(position_health_ratio_df, x='Which of the following best describes your work position?', 
             y='yes ratio',
             color='yes ratio'
             , height=600)
fig.show()
```
